```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
import uproot
```

```python
fname_in = '../smearing_matrices/smearing.root'
dir_out = '../smearing_matrices/'
```

```python
aux_file = uproot.open(fname_in)
```

```python
aux_npy = {}
for key, item in aux_file.items():
    print(key)
    np.save(dir_out + key.replace(';1', '') + '_matrix.npy', item.to_numpy()[0])
    np.save(dir_out + key.replace(';1', '') + '_binning_true.npy', item.to_numpy()[1])
    np.save(dir_out + key.replace(';1', '') + '_binning_reco.npy', item.to_numpy()[2])
    plt.figure()
    plt.pcolor(item.to_numpy()[0])
```
